## Solution

````python
import os
import pandas as pd

import bokeh.models
import bokeh.plotting
import bokeh.io

bokeh.io.output_notebook()
````

**a)** Load in the data set using `pd.read_csv()`.

````python
data_path = '../data'

df = pd.read_csv(os.path.join(data_path, 'penguins_subset.csv'), header=[0, 1])

# Take a look
df.head()
````

````python
!head ../data/penguins_subset.csv
````

**b)** This is not tidy. It violates the one-row-one-observation rule.

**c)** 

````python
df_tidy = df.stack(
    level=0
).sort_index(
    level=1
).reset_index(
    level=1
).rename(
    columns={"level_1": "species"}
)

# Takea  look
df_tidy.head()
````

````python
df_tidy.melt(id_vars='species', var_name='quantity', value_name='value')
````

Yes, this is tidy.

**d)**

````python
# Slice them out as a Numpy array.
gentoo_bill_lengths = df_tidy.loc[df_tidy["species"] == "Gentoo", "bill_length_mm"].values

# Takea  look
gentoo_bill_lengths
````

**e)**

````python
df_tidy.groupby('species')
````

````python
df_summary = df_tidy.groupby('species').mean()
df_summary['body_mass_kg'] = df_summary['body_mass_g'] / 1000.0

# Take a look
df_summary
````

Note that I don't care that I have an extra column with the body mass in grams. It does not hurt anything with a data set this small.

## Exercise 2.2

Make a scatter plot of bill length versus flipper length with the glyphs colored by species.

````python
import colorcet

colors = colorcet.b_glasbey_category10
````

````python
# Colors I would like to use
colors = colorcet.b_glasbey_category10

colors = ['#66c2a5','#fc8d62','#8da0cb']


# Instantiate figure
p = bokeh.plotting.figure(
    frame_height=200,
    frame_width=200,
    x_axis_label="flipper length (mm)",
    y_axis_label="bill length (mm)",
    toolbar_location='above',
)


# Populate glyphs
glyphs = []
species = []
for i, (s, sub_df) in enumerate(df_tidy.groupby('species')):
    species.append(s)
    glyphs.append(
        p.circle(
            sub_df['flipper_length_mm'], 
            sub_df['bill_length_mm'],
            color=colors[i], 
        )
    )

# Build a legend (necessary if we want it outside plot area)
legend = bokeh.models.Legend(
    items=[(s, [g]) for s, g in zip(species, glyphs)],
    location="right",
    click_policy="hide",
)

# Add legend to plot
p.add_layout(legend, "right")

bokeh.io.show(p)
````
